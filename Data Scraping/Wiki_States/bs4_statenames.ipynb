# Import Modules

```python
import re
from bs4 import BeautifulSoup
import requests
import pandas as pd
from IPython.display import display
```

# Get the URL of the page

```python
get_url = requests.get("https://en.wikipedia.org/wiki/List_of_sovereign_states")
soup = BeautifulSoup(get_url.content, "html.parser")
```

# Extract the table

```python
find_table = soup.find("table", class_=re.compile("sortable wikitable"))
list_rows = find_table.find_all("tr")[1:]

state_names = []
for row in list_rows:
    # print(row)
    cols = row.find_all("td")
    if cols:
        state_name = [col.get_text(strip=True) for col in cols[:4]]
        state_names.append(state_name)
```

# Convert to Dataframe, Export to CSV and JSON, and display the data

```python
pd.set_option('display.max_rows', None)  

df = pd.DataFrame(state_names, columns=["State Names", "UN Membership", "Sovereignty", "More Information"])
df.to_csv("wiki_states.csv", index=Falscsve)
df.to_json("wiki_states.json", orient="records", lines=True)
print(f"{df.shape[0]} rows and {df.shape[1]} columns")
display(df.iloc[:])
```
